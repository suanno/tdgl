```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
ts = np.array([125/100, 3, 10, 41])    # Time for 10 steps
Ns = np.array([512, 1024, 2048, 4096])

fig, ax = plt.subplots()
ax.plot(np.log(Ns), np.log(ts/9),'.')
c = np.polyfit(np.log(Ns), np.log(ts/9), deg=1)
x_fit = np.linspace(min(np.log(Ns)),max(np.log(Ns)),1000)
y_fit = c[0]*x_fit  + c[1]*np.ones(len(x_fit))
ax.plot(x_fit,y_fit,'-')
ax.set_xlabel("log N")
ax.set_ylabel("log computational time")
ax.set_title("Time for one integration step")
print(c[0])
print(np.exp(c[1]))
```
