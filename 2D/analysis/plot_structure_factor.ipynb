# Structure factor (2D)
$$\hat{S}(qx,qy)\equiv |\mathcal{FFT}(u(x))|^2$$
where, in fftw, both the direct and the inverse FFT do not have normalization (you must normalize the by 1/N **manually** when you do the IFT to revert the trasformation)

THEN we evaluate this function only along two horizontal lines passing through the origin. And we take the average. So we are left with a function of only one variable.
