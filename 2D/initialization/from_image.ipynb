```python
import numpy as np
import scipy
from matplotlib import pyplot as plt
```

```python
import os, shutil
name = "15_10_B"
directory = "../.saves/"+str(name)
if not os.path.exists(directory):
    os.makedirs(directory)

shutil.copyfile("../params.txt", directory+"/params.txt")
```

```python
from PIL import Image
L = 128
dx = 0.1
N = int(L/dx)
img = Image.open('.images/io.png')
img = img.convert('1')
img = img.resize((N, N), Image.ANTIALIAS) # resizes image in-place
#lum_img = image[:,:,0]
img.save(".images/temp.bmp")

f = open(directory+"/state.dat", "w")
f.write(str(N)+" 0 "+str(dx)+"\n")
f.close()
f = open(directory+"/state.dat", "a")
image = plt.imread(".images/temp.bmp")
dx = 0.1
x = 0
y = 0
for i in np.arange(0,N):
    y = 0
    for j in np.arange(0,N):
        if image[i,j,0] == 0:
            f.write(str(x)+" "+str(y)+" "+str(-1)+"\n")
        else:
            f.write(str(x)+" "+str(y)+" "+str(1)+"\n")
        y = round(y + dx, 3)
    x = round(x + dx,3)
f.close()
```
