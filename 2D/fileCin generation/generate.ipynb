```python
import os
import shutil
import subprocess
import time
import numpy as np
from matplotlib import pyplot as plt
```

```python
dt = 0.1
tspan = 10000
N = int(tspan/dt)
Cs = np.zeros(N)
ts = np.linspace(0,tspan,N)
```

```python
# Create evolution directory
simul_name = "9_09_D"

file_simul = ".saves/" + simul_name + "/" 
name = "fileCin.dat"
dir = "../" + file_simul + name
# Copy files
f = open(dir, "w")
T = 40
for i in np.arange(0,N):
    Cs[i] = 1 + 0.8*np.sin(2*np.pi*ts[i]/T)
    """
    if i*dt < 5:
        Cs[i] = -1
    else:
        Cs[i] = 1
    #Cs[i] = -0.5
    """
    f.write(str(ts[i]) + " " + str(Cs[i]) + "\n")
f.close()

plt.plot(ts, Cs)
plt.title("dt = " + str(dt))
plt.ylabel("C(t)")
plt.xlabel("t")
```
