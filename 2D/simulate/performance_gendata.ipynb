```python
import numpy as np
from matplotlib import pyplot as plt
import os, time
```

```python
dx = 0.1
numsteps = 100
tspan = 1
```

```python
# Create a temporary folder with fileCin.dat file inside
temp_folder_name = "temp"
ts = np.linspace(0,2*tspan,2*numsteps)
Cs = np.ones(len(ts))*1
f = open("../.saves/"+temp_folder_name+"/fileCin.dat", "w")
for i in np.arange(0,len(ts)):
    f.write(str(ts[i]) + " " + str(Cs[i]) + "\n")
f.close()
```

```python
Ns = np.array([4096])

NumSamples = 10 # Number of samples per each simulation
```

```python
dir = "computational_times/numsteps="+str(numsteps)+"/"
if not os.path.isdir(dir):
    os.makedirs(dir)
for N in Ns:
    time_file = dir + str(N) + ".dat"
    # Simulations
    f = open(time_file,"w")
    for i in np.arange(NumSamples):
        os.system("cd ..; ./.bin/datainit "+str(N)+" "+str(dx)+ " 'temp'")
        t = time.time()
        os.system("cd ..; ./.bin/tdgl "+str(tspan)+" 'temp'")
        elapsed = time.time() - t
        f.write(str(elapsed)+"\n")
    f.close()
```

```python
dir = "save_computational_times/numsteps="+str(numsteps)+"/"
if not os.path.isdir(dir):
    os.makedirs(dir)
for N in Ns:
    time_file = dir + str(N) + ".dat"
    # Simulations
    f = open(time_file,"w")
    for i in np.arange(NumSamples):
        os.system("cd ..; ./.bin/datainit "+str(N)+" "+str(dx)+ " 'temp'")
        t = time.time()
        os.system("cd ..; ./.bin/test_save 0 'temp'")
        elapsed = time.time() - t
        f.write(str(elapsed)+"\n")
    f.close()
```
