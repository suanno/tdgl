```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Double well potential

def V(x):
    return -0.5*x**2 + 0.25*x**4
def parabola(x):
    return 0.1*x**2

N = 1000
xs = np.linspace(-1.5, 1.5, N)
ys = np.zeros(N)
ys_ = np.zeros(N)
for i in np.arange(0, N):
    ys[i] = V(xs[i])
    ys_[i] = parabola(xs[i])

fig, ax = plt.subplots()

ax.plot(xs, ys, label = "T < Tc")
ax.plot(xs, ys_, label = "T > Tc")
ax.set_title("Double well potential V(x)")
ax.set_xlabel("x")
ax.set_ylabel("V(x)")
ax.set_axis_off()
ax.legend()
```

```python
# Stationary states
N = 1000
L = 10
x = np.linspace(0,L,N)
fig, ax = plt.subplots(1,2)
y = np.ones(N)
ax[0].plot(x,y,label = "m = +1")
ax[0].plot(x,-y,label = "m = -1")
for i in np.arange(0, N):
    y[i] = np.tanh(x[i]-0.5*L)
ax[1].plot(x,y)
ax[0].legend()

ax[0].set_xlabel("x")
ax[0].set_ylabel("m(x)")
ax[1].set_xlabel("x")

#ax[0].set_axis_off()
```
