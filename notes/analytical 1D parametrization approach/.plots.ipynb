```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def alpha(t, a, alpha_0):
    t0 = 1/(2*a*alpha_0**2)
    if a > 0:
        if t > t0:
            return np.nan
    else:
        if t < t0:
            return np.nan
    return np.sign(alpha_0)/np.sqrt(2*a*(t0-t))
def beta(t, a, b, beta_0):
    t0 = 1/(2*a/b*beta_0**2)
    if a > 0:
        if t > t0:
            return np.nan
    else:
        if t < t0:
            return np.nan
    beta_t = np.sign(beta_0)/np.sqrt(2*a/b*(t0-t))
    return beta_t
def C(t, a, b, t0):
    if a > 0:
        if t > t0:
            return np.nan
    else:
        if t < t0:
            return np.nan 
    return -(1+b/a)/(2*(t-t0))
```

```python
# Plot \alpha(t)

fig, ax = plt.subplots()
ax.set_xlabel("t")

N = 100

a = 1
alpha_0 = 1
t0 = 1/(2*a*alpha_0**2)

ts = np.linspace(-5,5,N)
alphasp = np.zeros(N)
alphasm = np.zeros(N)
for i in np.arange(0,N):
    alphasp[i] = alpha(ts[i], a, alpha_0)
    alphasm[i] = alpha(ts[i], a, -alpha_0)

ax.plot(ts, alphasm, label="alpha_0 < 0")
ax.plot(ts, alphasp, label="alpha_0 > 0")

ax.set_title("a > 0\n alpha(t)")
ax.set_xlim(-5,5)
ax.axvline(t0, color="black", label = "t0 > 0")
ax.legend()
```

```python
# Plot C(t)

fig, ax = plt.subplots()
ax.set_xlabel("t")

N = 100

a = -1
b = 0.5
alpha_0 = 1
t0 = 1/(2*a*alpha_0**2)

ts = np.linspace(-5,5,N)
Csp = np.zeros(N)
Csm = np.zeros(N)
for i in np.arange(0,N):
    Csm[i] = C(ts[i], a, b, t0)
    Csp[i] = C(ts[i], a, b - a, t0)

ax.plot(ts, Csm, label="0 < b < -a")
ax.plot(ts, Csp, label="b > -a")

ax.set_title("a < 0\n C(t)")
ax.set_xlim(-5,5)
ax.axvline(t0, color="black", label = "t0 < 0")
ax.legend()
```

```python
# Double well potential

def V(x):
    return 0.5*x**2 - 0.25*x**4

N = 1000
xs = np.linspace(-1.5, 1.5, N)
ys = np.zeros(N)
for i in np.arange(0, N):
    ys[i] = V(xs[i])

plt.plot(xs, ys)
plt.title("Double well potential")
plt.xlabel("x")
plt.ylabel("V(x)")
```
