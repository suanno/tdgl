### Here we monitor the average of q2 and on the same time the one of u.
That's because we want to see how <q2> scales **in the linear regime**, so when u is small everywhere (and its average is a measure of how small it is)

```python
f1 = open("../../fileAveout.dat", "r")
f2 = open("../../fileq2Aveout.dat", "r")
f3 = open("../../fileumax.dat", "r")
N = 1000000000
tt = np.zeros(N)
uave = np.zeros(N)
q2ave = np.zeros(N)
umax = np.zeros(N)
i = 0
for row in f1:
    a, b = row.split(" ")
    tt[i] = float(a)
    uave[i] = float(b)
    i = i + 1
uave = uave[0:i]
tt = tt[0:i]
i = 0
for row in f2:
    a, b = row.split(" ")
    #tt[i] = float(a)
    q2ave[i] = float(b)
    i = i + 1
q2ave = q2ave[0:i]
i = 0
for row in f3:
    a, b = row.split(" ")
    #tt[i] = float(a)
    umax[i] = float(b)
    i = i + 1
umax = umax[0:i]

fig, ax = plt.subplots()
ax.plot(tt, uave)
ax.set_title("<u>")

fig, ax = plt.subplots()
ax.loglog(tt, q2ave)
ax.set_title("<q2>")

fig, ax = plt.subplots()
ax.plot(tt, umax)
ax.set_title("umax")
```
