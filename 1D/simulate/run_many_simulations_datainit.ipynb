```python
import numpy as np
from matplotlib import pyplot as plt
import os
import shutil
```

```python
tspan = 10000
tspan_ini = 10  # To tune depending on A, C, T (I expect it is not dependent on L, dt, dx [for sufficiently small values of them])            
dt = 0.1
dt_ini = 0.001

L = 2048
dx = 0.1

continue_simulation = 1
```

```python
def delete_files_main_dir():
    file_list = ["fileq2Aveout.dat", "fileCout.dat", "fileellDW.dat", "fileSq.dat"]
    for file in file_list:
        open("../"+file, "w").close()
        print("Empied: ../"+file)
```

```python
Cbar_list = np.array([0.01])
T_list = np.array([30])
A_list = np.array([1])

for T in T_list:
    for A in A_list:
        for Cbar in Cbar_list:
            delete_files_main_dir()
            time = 0
            # Define save folder for this set of parameters
            simul_name = "Cbar="+str(round(Cbar,3))+"/T="+str(round(T,3))+"/A="+str(round(A,3))+"/"
            dir_save = "symmetric/L="+str(L)+"/" + simul_name
            dir = "../.saves/" + dir_save
            # Check if there is already a simulation, with the specified value of parameters, that can be continued
            if continue_simulation == 1:
                if not os.path.isdir(dir):
                    continue_simulation = 0
            
            if continue_simulation == 0:
                # Prepare the state (twokinks)
                N = int(L/dx)
                L = N*dx
                #d0 = 20
                u0 = np.sqrt(Cbar)

                os.system("cd ..; ./.bin/datainit "+str(N)+" "+str(0.1)) 
                if not os.path.isdir(dir):
                    os.makedirs(dir)
            else:
                # Copy state of the system into the work directory (to evolve it further)
                shutil.copyfile(dir + "fileCout.dat", "../fileCout.dat")
                shutil.copyfile(dir + "fileq2Aveout.dat", "../fileq2Aveout.dat")
                shutil.copyfile(dir + "tdgl_result.dat", "../tdgl_result.dat")
                shutil.copyfile(dir + "parameters.dat", "../parameters.dat")
                shutil.copyfile(dir + "fileellDW.dat", "../fileellDW.dat")

            """Run simulation"""
            if continue_simulation == 0:
                # Initial dynamics (its duration depends on 1/C)
                dtspan_ini = min(1000*dt_ini, tspan_ini)
                while time < tspan_ini:
                    os.system("cd ..; ./.bin/tdgl "+str(dtspan_ini)+" "+str(A)+" "+str(T)+" "+str(Cbar)+" 0 "+str(dt_ini))
                    time = time + dtspan_ini
            # Asymptotic dynamics
            dtspan = 1000*dt
            while (time < tspan):
                os.system("cd ..; ./.bin/tdgl "+str(dtspan)+" "+str(A)+" "+str(T)+" "+str(Cbar)+" 0 "+str(dt))
                time = time + dtspan
            
            # Save
            shutil.copyfile("../fileCout.dat", dir + "fileCout.dat")
            shutil.copyfile("../fileq2Aveout.dat", dir + "fileq2Aveout.dat")
            shutil.copyfile( "../fileellDW.dat", dir + "fileellDW.dat")
            shutil.copyfile( "../fileSq.dat", dir + "fileSq.dat")
            shutil.copyfile("../tdgl_result.dat", dir + "tdgl_result.dat")
            shutil.copyfile("../parameters.dat", dir + "parameters.dat")
            print("Saved at "+dir)
```
